~Importing datset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("Data_Train.csv")
```

```python
df.head()
```

Data cleaning

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Airline'].value_counts()
```

```python
df['Source'].unique()
```

```python
df['Destination'].unique()
```

```python
df['Day'] = pd.to_datetime(df.Date_of_Journey, format="%d/%m/%Y").dt.day
```

```python
df['Month'] = pd.to_datetime(df.Date_of_Journey, format="%d/%m/%Y").dt.month
```

```python
df.sample(5)
```

```python
df.drop(['Date_of_Journey'], axis =1, inplace=True)
```

```python
df['Dep_Time'] = pd.to_datetime(df.Dep_Time).dt.hour
```

```python
df['Arrival_Time'] = pd.to_datetime(df.Arrival_Time).dt.hour
```

```python
def categorize_time(hour):    
    if 5 <= hour < 12 :
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'
df['Dep_Time'] = df['Dep_Time'].apply(categorize_time)
df['Arrival_Time'] = df['Arrival_Time'].apply(categorize_time)
```

```python
df.sample(5)
```

```python
def convert_to_minutes(duration):
    hours = 0
    minutes = 0
    if 'h' in duration:
        hours_part = duration.split('h')[0]
        hours = int(hours_part)
        minutes_part = duration.split('h')[1].replace('m', '')
        if minutes_part:
            minutes = int(minutes_part)
    else:
        minutes = int(duration.replace('m', ''))
        
    total_minutes = hours*60+minutes
    return total_minutes
df['Duration'] = df['Duration'].apply(convert_to_minutes)    
```

```python
df.sample(5)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].replace({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4})
df.head()
```

```python
df.drop(['Route', 'Additional_Info'],axis =1, inplace = True)
```

```python
df.sample(3)
```

Remove Outliers

```python
df[['Price','Duration']].describe()
```

```python
def remove_outliers(df, column):
    mean = df[column].mean()
    std = df[column].std()
    return df[(df[column]<= mean-3*std) & (df[column]<= mean+3*std)]
df_no_outliers = remove_outliers(df, 'Price') 
```
